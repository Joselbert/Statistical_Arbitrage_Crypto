```python
import joblib 
    
def bestReturn(name_file, input_folder):
    name = name_file.split(".")[0]
    res = joblib.load(f"{input_folder}/{name_file}")
    columns = ["name", "month", "maximum_return","trading_window", "p_value"]
    names = []
    months = []
    maximum_returns = []
    trading_window = []
    p_value = []
    for k in res.keys():
        mat = res[k][0]
        mat = mat.astype(float)
        col = mat.columns
        ind = mat.index
        ind_, col_ = np.where(mat.to_numpy() == mat.to_numpy().max())
        max_date = ind[ind_.item()]
        max_pval = col[col_.item()]
        max_val = mat.to_numpy().max()
        month = k

        names.append(name)
        months.append(month)
        maximum_returns.append(max_val)
        trading_window.append(max_date)
        p_value.append(max_pval)

    df = pd.DataFrame(columns = columns )
    df["name"] = names
    df["month"] = months
    df["maximum_return"] = maximum_returns
    df["trading_window"] = trading_window
    df["p_value"] = p_value
    
    return df


def worstReturn(name_file, input_folder):
    name = name_file.split(".")[0]
    res = joblib.load(f"{input_folder}/{name_file}")
    columns = ["name", "month", "minimum_return","trading_window", "p_value"]
    names = []
    months = []
    minimum_returns = []
    trading_window = []
    p_value = []
    for k in res.keys():
        mat = res[k][0]
        mat = mat.astype(float)
        col = mat.columns
        ind = mat.index
        ind_, col_ = np.where(mat.to_numpy() == mat.to_numpy().min())
        min_date = ind[ind_.item()]
        min_pval = col[col_.item()]
        min_val = mat.to_numpy().min()
        month = k

        names.append(name)
        months.append(month)
        minimum_returns.append(min_val)
        trading_window.append(min_date)
        p_value.append(min_pval)

    df = pd.DataFrame(columns = columns )
    df["name"] = names
    df["month"] = months
    df["minimum_return"] = minimum_returns
    df["trading_window"] = trading_window
    df["p_value"] = p_value
    
    return df
```
